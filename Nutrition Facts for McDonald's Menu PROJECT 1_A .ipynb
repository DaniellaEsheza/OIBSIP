## Nutrition Facts for McDonald's Menu

##### This repository provides detailed nutritional information for items on the McDonald's menu. 
##### It includes calories, fats, carbohydrates, proteins, and other key nutritional metrics for each item.

### Importing Required Libraries

### Calculating Specific Statistics for Calories

```python
mean_calories = menu['Calories'].mean()
median_calories = menu['Calories'].median()
mode_calories = menu['Calories'].mode()[0]
std_calories = menu['Calories'].std()


print(f"Mean Calories: {mean_calories}")
print(f"Median Calories: {median_calories}")
print(f"Mode Calories: {mode_calories}")
print(f"Standard Deviation of Calories: {std_calories}")
```

### Grouping and Analyzing Average Calories per Category

```python
category_calories = menu.groupby('Category')['Calories'].mean().sort_values()

plt.figure(figsize=(15, 5))
category_calories.plot(kind='bar', color='green')
plt.title('Average Calories per Category')
plt.xlabel('Category')
plt.ylabel('Average Calories')
plt.xticks(rotation=45)
plt.show()
```

### Identifying Top 5 High Calorie Items/High Protein Items

```python
top_caloric_items = menu[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(5)
print("Top 5 highest calorie items:\n", top_caloric_items)

top_protein_items = menu[['Item', 'Protein']].sort_values(by='Protein', ascending=False).head(5)
print("Top 5 highest protein items:\n", top_protein_items)

plt.figure(figsize=(12, 7))
plt.bar(top_caloric_items['Item'], top_caloric_items['Calories'], color='grey')
plt.title('Top 5 High Calorie Items')
plt.xlabel('Item')
plt.ylabel('Calories')
plt.xticks(rotation=45)
plt.show()
```

### Plotting a Histogram of Calorie Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(menu['Calories'], bins=30, kde=True, color='teal')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

### Correlation Heatmap for Numeric Columns

```python
numeric_df = menu.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(20, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Nutritional Correlation Heatmap')
plt.show()
```

### Barplot of Protein vs. Calories

```python

plt.figure(figsize=(15, 5))
sns.barplot(x='Protein', y='Calories', data=menu, hue='Protein', legend=False, palette='viridis', errorbar=None)
plt.title('Calories vs Protein')
plt.show()
```
